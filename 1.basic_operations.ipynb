# Basic operations on pointclouds

In this notebook we discuss the basic operations on pointclouds.

```python
# Necessary imports
import numpy as np
import matplotlib.pyplot as plt
```

## What is a pointcloud?

In general terms, a pointcloud refers to a set of data points in space (this space may have multiple and unhomogeneous dimensions such as speed, temperature, color, etc.). We assume here a three-dimensional space where **a pointcloud is a representation of a shape or an object** (or pieces of one or several objects).

Pointclouds may be obtained from real objects (through "3D scanning" processes), from digital objects (through "3D modeling" processes), or even created from scratch. This results in a multiplication of available data, but also formats, each one being tailored for specific uses and software.

For the sake of simplicity, we are going to use a simple ASCII-based format (i.e., a plain text file) in this notebook. Note that binary formats (such as PLY, LAS, or E57) must be privileged for large pointclouds.

A pointcloud is then simply loaded using Numpy:

```python
points = np.loadtxt("./data/standford_bunny.xyz")
print(points)
```

We can see that pointcloud is **basically a "list" of points** (the notion of set is more appropriate) defined by their cartesian coordinates (i.e., x, y, and z).

We can easily visualize them using Matplotlib if their number is not too large (otherwise other libraries or specific applications may be needed).

Note the **sparse** (some areas of the object may not be sampled), **irregular** (some areas may have different densities), and **unordered** (the "order" of points is not important) nature of pointclouds.

```python
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection="3d")
ax.scatter(points[:, 0], points[:, 1], points[:, 2], color="brown")
ax.view_init(10, 60)
plt.axis("equal")
plt.show()
```
