# Basic operations on pointclouds

In this notebook we discuss the basic operations on pointclouds.

```python
# Necessary imports
import numpy as np
import matplotlib.pyplot as plt
```

## What is a pointcloud?

In general terms, a pointcloud refers to a set of data points in space (this space may have multiple and unhomogeneous dimensions such as speed, temperature, color, etc.). We assume here a three-dimensional space where **a pointcloud is a representation of a shape or an object** (or pieces of one or several objects).

Pointclouds may be obtained from real objects (through "3D scanning" processes), from digital objects (through "3D modeling" processes), or even created from scratch. This results in a multiplication of available data, but also formats, each one being tailored for specific uses and software.

For the sake of simplicity, we are going to use a simple ASCII-based format (i.e., a plain text file) in this notebook. Note that binary formats (such as PLY, LAS, or E57) must be privileged for large pointclouds.

A pointcloud is then simply loaded using Numpy:

```python
points = np.loadtxt("./data/standford_bunny_simple.xyz")
print("Here is the content of our pointcloud:\n", points)
```

We can see that **a pointcloud is basically a "list" of points** (the notion of set is more appropriate) defined by their cartesian coordinates (i.e., x, y, and z).

We can easily visualize them using Matplotlib if their number is not too large (otherwise other libraries or specific applications may be needed).

Note the **sparse** (some areas of the object may not be sampled), **irregular** (some areas may have different densities), and **unordered** (the "order" of points is not important) nature of pointclouds.

```python
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(projection="3d")
ax.scatter(points[:, 0], points[:, 1], points[:, 2])
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(10, 60)
plt.axis("equal")
plt.show()
```

## Common attributes

Each point of the pointcloud has *a minima* three coordinates (x, y, and z), but it is not uncommon to find other informations. The most popular ones are normals, colors, and scalar fields.


```python
data = np.loadtxt("./data/standford_bunny_custom.xyz")
print("Here is the shape of our data:", data.shape)
points = data[:, :3]
colors = data[:, 3:6]
scalar = data[:, 6]
normals = data[:, 7:]
```

**A normal is a vector that is assumed to be perpendicular to the sampled surface** at a given point. 

A normal vector has three components, noted nx, ny, and nz. It is generally a unit vector, which means that its norm is equal to one.

Normals are sometimes given directly by the acquisition device or may be easily calculated when the sampled surface is known (if a pointcloud comes from a CAD model for example). When this is not the case, normals can be estimated by various methods (we'll get to it in a next notebook).

```python
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection="3d")
ax.quiver(points[:, 0], points[:, 1], points[:, 2],
          5.*normals[:, 0], 5.*normals[:, 1], 5.*normals[:, 2]) # we make normals a little bit larger for visualization purposes
ax.view_init(10, 60)
ax.set_axis_off()
plt.axis("equal")
plt.show()
```

Normals provide useful local information about the geometrical properties of the sampled surface or object. In particular, normals help define the notion of "interior/exterior" (normals are supposed to be pointing "outside" by convention).

Normals are used as inputs in many pointcloud processing algorithms, but also for visualization/rendering purposes. Computing the angle between an incident ray and the normals is a first approach to simulating the reflexion of light on a given object.

```python
# Incident ray (unit vector giving the direction of light)
v_dir = np.array([-1., 0., 0.])
v_loc = points.mean(axis=0) - max(points @ v_dir) * v_dir # used for visualization purposes

# Angle between the normals and the incident ray
theta = np.arccos(normals @ v_dir)

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection="3d")
p = ax.scatter(points[:, 0], points[:, 1], points[:, 2],
          c=theta, cmap="bone")
ax.quiver(v_loc[0], v_loc[1], v_loc[2], 25.*v_dir[0], 25.*v_dir[1], 25.*v_dir[2], color='r') # direction of view
ax.view_init(10, 60)
ax.set_axis_off()
cbar = fig.colorbar(p, shrink=0.5)
cbar.set_label('angle (rad)')
plt.axis("equal")
plt.show()
```

Colors are more easily defined and account for the visual apparance of each point.

There are several ways to define colors, one of the most common is **the RGB color model that defines a color as an addition of the red, green and blue primary colors**.

A RGB color vector has three components, noted R, G, and B, with values generally ranging from 0 to 255 or from 0 to 1.

Colors are sometimes given by the acquisition device or may be easily deduced if the pointcloud comes from a colored virtual object. However, in the first case, color values may vary to the acquisition conditions (the lighting in particular).

```python
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection="3d")
ax.scatter(points[:, 0], points[:, 1], points[:, 2],
           c=colors/255.) # RGB values need to be normalized in Matplotlib
ax.view_init(10, 60)
ax.set_axis_off()
plt.axis("equal")
plt.show()
```

In general terms, a scalar field is a function associating a single number to every point in a space. Here it simply designates **a number associated with a point**.

This number may have any value and dimension.

Scalar fields are sometimes given by the acquisition device or may result of a calculation. In practice, a scalar field is used to associate any quantity of interest with the pointcloud.

```python
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection="3d")
p = ax.scatter(points[:, 0], points[:, 1], points[:, 2],
           c=scalar)
ax.view_init(10, 60)
ax.set_axis_off()
cbar = fig.colorbar(p, shrink=0.5)
cbar.set_label('point quality')
plt.axis("equal")
plt.show()
```

## Transformations

Many kinds of geometric transformations may be applied to pointclouds (their study from a mathematical perspective is of course beyond the scope of this discussion). In the context of pointcloud processing, the following transformations are mainly considered (alone or combined):
- Translations
- Rotations
- Reflections
- Scaling
- Shearing

These transformations may roughly be classified according to the properties of the initial pointcloud that are preserved. For example, rigid motions (or Euclidean transformations or isometries) - that include translations, rotations and reflections - preserve angles and distances between points. On the other hand, affine transformations - that include rigid motions plus scaling and shearing - do not.

### Translations

A translation moves every point of the pointcloud in the same direction.

It is implemented in practice by adding a constant vector to each point of the pointcloud (thanks to broadcasting this operation may be vectorized in Numpy, resulting in more readable code and faster execution).

A translation may be used to "center" a pointcloud, for example to make it roughly coincide with the origin of the coordinate system or with another pointcloud. This allows to introduce the notion of centroid (or center of mass if all points are weighted equally), which is the mean position of all the points of the pointcloud. This notion is widely used in pointcloud processing algorithms.

```python
# Translated points
translation = np.array([50., 40., -10.])
translated_points = points + translation
# Centered points
centroid = points.mean(axis=0)
centered_points = points - centroid

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection="3d")
ax.scatter(points[:, 0], points[:, 1], points[:, 2], c='r')
ax.scatter(translated_points[:, 0], translated_points[:, 1], translated_points[:, 2], c='g')
ax.scatter(centered_points[:, 0], centered_points[:, 1], centered_points[:, 2], c='b')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(10, 60)
plt.axis("equal")
plt.show()
```

### Rotations

A rotation moves every point of the pointcloud around a certain axis.

Elementary rotation matrices for a fixed coordinate system.

\begin{equation*}
R_{x}(\alpha) = 
\begin{pmatrix}
1 & 0 & 0 \\
0 & \cos\alpha & -\sin\alpha \\
0 & \sin\alpha & \cos\alpha
\end{pmatrix}
\quad
R_{y}(\beta) = 
\begin{pmatrix}
\cos\beta & 0 & \sin\beta \\
0 & 1 & 0 \\
-\sin\beta & 0 & \cos\beta
\end{pmatrix}
\quad
R_{z}(\gamma) = 
\begin{pmatrix}
\cos\gamma & -\sin\gamma & 0 \\
\sin\gamma & \cos\gamma & 0 \\
0 & 0 & 1 
\end{pmatrix}
\end{equation*}

Rotation using an given a unit vector $u$ and angle $\theta$

\begin{equation*}
R_{u}(\theta) = 
\begin{pmatrix}
u_{x}^{2}(1-c)+c & u_{x}u_{y}(1-c)-u_{z}s & u_{x}u_{z}(1-c)+u_{y}s \\
u_{x}u_{y}(1-c)+u_{z}s & u_{y}^{2}(1-c)+c & u_{y}u_{z}(1-c)-u_{x}s \\
u_{x}u_{z}(1-c)-u_{y}s & u_{y}u_{z}(1-c)+u_{x}s & u_{z}^{2}(1-c)+c 
\end{pmatrix}
\quad
\text{with}
\quad
c = \cos\theta
\quad
\text{and}
\quad
s = \sin\theta
\end{equation*}

Composition of rotations
\begin{equation*}
R_{t} = R_{x}(\alpha).R_{y}(\beta).R_{z}(\gamma)
\end{equation*}
